```python
img1 = cv2.imread('giraffes.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_RGB2BGR)
img2 = cv2.imread('bricks.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_RGB2BGR)
```

```python
import matplotlib.pyplot as plt
```

resize images

```python
img1 = cv2.resize(img1,(1400,1400))
img2 = cv2.resize(img2,(1400,1400))
```

```python
plt.imshow(img2)
```

```python
blended = cv2.addWeighted(src1=img1,alpha=0.5,src2=img2,beta=0.9,gamma=0)
```
