```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    img = cv2.imread('bricks.jpg').astype(np.float32) / 255
    img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
    return img
```

```python
# load_img()
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
i = load_img()
display_img(i)
```

```python
gamma = 4
```

```python
result = np.power(i,gamma)
display_img(result)
```
